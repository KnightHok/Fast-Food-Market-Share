```python
import pandas as pd
import numpy as np
from datetime import datetime, timezone
```

```python
df = pd.read_csv("apify_export.csv")
df
```

```python
cols = ["bizId", "name", "aboutTheBusiness/yearEstablished", "aggregatedRating", "address/addressLine1", "address/addressLine2", "address/city", "address/country", "address/postalCode", "address/regionCode", "categories/0", "categories/1", "categories/2", "healthScore", "cuisine", "directUrl"]
new_df = df[cols].copy()
new_df
```

```python
mask = (new_df["categories/0"] == "Fast Food") | (new_df["categories/1"] == "Fast Food") | (new_df["categories/2"] == "Fast Food")
new_df["keep"] = np.where(mask, 1, 0)
```

```python
df_filtered = new_df[new_df["keep"] == 1].copy()
df_filtered = df_filtered.drop(columns=["keep"])
df_filtered
```

```python
export_time = datetime.now(timezone.utc).isoformat()
df_filtered.to_csv(f"exports/run_{export_time}.csv", index=False)
```
